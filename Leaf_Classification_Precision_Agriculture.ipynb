#Loading Dataset from hugging face

```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import models, layers
import pandas as pd
```

```python
train_dataset = load_dataset("yusuf802/leaf-images", split="train")
test_dataset = load_dataset("yusuf802/leaf-images", split="test")
```

```python
label_names = train_dataset.features["label"].names
def extract_species_from_label(name):
    # try a few separators; adjust if your labels use a different convention
    return name.split('_')[0]

# build species list from label_names
species_names = sorted({extract_species_from_label(str(n)) for n in label_names})
print(species_names)
print(len(species_names))
```

# Preprocess

```python
from PIL import Image
import numpy as np
import tensorflow as tf

IMG_SIZE = 128  # resize to 128x128

def preprocess(example):
    img = example["image"].convert("RGB")       # ensure 3 channels
    img = img.resize((IMG_SIZE, IMG_SIZE))      # resize
    img = np.array(img) / 255.0                 # normalize
    return {"image": img, "label": example["label"]}

train_dataset = train_dataset.map(preprocess)
test_dataset  = test_dataset.map(preprocess)
```

#Build a Simple Transfer Learning Model (MobileNetV2)

```python
base = tf.keras.applications.MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3), include_top=False, weights="imagenet", pooling="avg"
)
base.trainable = False

model = models.Sequential([
    tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    base,
    layers.Dropout(0.3),
    layers.Dense(len(label_names), activation="softmax")
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
model.summary()
```

```python
BATCH_SIZE = 32

train_tf = train_dataset.to_tf_dataset(
    columns=["image"],
    label_cols=["label"],
    shuffle=True,
    batch_size=BATCH_SIZE,
)

test_tf = test_dataset.to_tf_dataset(
    columns=["image"],
    label_cols=["label"],
    shuffle=False,
    batch_size=BATCH_SIZE,
)
```

#Train Quickly (Beginner-Friendly)

```python
from tensorflow.keras.callbacks import EarlyStopping

early = EarlyStopping(monitor="val_accuracy", patience=3, restore_best_weights=True)

history = model.fit(train_tf, validation_data=test_tf, epochs=5, callbacks=[early])
```

Evaluate and Save the Model

```python
loss, accuracy = model.evaluate(test_dataset)
print(f"Test accuracy: {accuracy*100:.2f}%")

model.save("leaf_species_model")
import json
with open("species_labels.json", "w") as f:
    json.dump(label_names, f)
```
